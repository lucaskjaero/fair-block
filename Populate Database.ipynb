```python
BASE_URL = "https://blockchain-team-badass.herokuapp.com"
PARKING_SPOT_URL = BASE_URL + "/parkingspots/"
STATE_URL = BASE_URL + "/state/"
```

```python
import pandas as pd

data = pd.read_table("parking_spots.csv", sep="\t")
```

```python
data.head()
```

```python
unique_parking_spots = data.drop_duplicates(subset="pole")
unique_parking_spots
```

```python
import json
import requests

headers = {'Content-type': 'application/json'}

def format_time(time_string):
    if "am" in time_string:
        return int(time_string.replace("am", ""))
    else:
        return int(time_string.replace("pm", "")) + 12

for index, spot in unique_parking_spots.iterrows():
    split_config = spot['config_name'].split(" ")
    max_hours = split_config[0]
    hour = split_config[3]
    enforced_time = split_config[5]
    days_enforced = split_config[6]
    
    hour = hour.replace("$", "").replace(" HR", "")

    start_enforcing, stop_enforcing = enforced_time.split("-")
    
    
    body = {
        "uuid": spot['pole'],
        "latitude": spot['latitude'],
        "longitude": spot['longitude'],
        "name": spot['sub_area'],
        "max_hours": max_hours,
        "start_enforcing": format_time(start_enforcing),
        "stop_enforcing": format_time(stop_enforcing),
        "days_enforced": days_enforced,
        "price": str(hour),
    }
    data_json = json.dumps(body)
    
    try:
        #print(body)
        response = requests.post(PARKING_SPOT_URL, data=data_json, headers=headers)
        print(response.text)
    except Exception as e:
        print(e)
```

```python
def convert_timestamp(old_stamp):
    day, time = old_stamp.split("_")
    month, day, year = day.split("/")
    hour, minute = time.split(":")
    
    year = "20" + str(year)
    
    return "%s-%s-%sT%s:%s:%sZ" % (year, month, day, hour, minute, "00")

headers = {'Content-type': 'application/json'}

for index, state in data.iterrows():
        
    body = {
        "parking_spot": state['pole'],
        "time_in": convert_timestamp(state['start_time_stamp']),
        "time_out": convert_timestamp(state['end_time_stamp']),
        "occupied_by": "aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa"
    }
    data_json = json.dumps(body)
    
    try:
        response = requests.post(STATE_URL, data=data_json, headers=headers)
        print(response.text)
    except Exception as e:
        print(e)
```
